Import justices from the SCOTUS webpage.

```python
from bs4 import BeautifulSoup
import os
import sys
import pandas as pd
from datetime import datetime
```

```python
with open(os.path.expanduser('~')+'/Dropbox/Research/py_lib/data_sets/scotus/Members of the Supreme Court of the United States - Text Version.html','r') as f:
    soup = BeautifulSoup(f)

justiceTable = soup.find_all('table')[1]
df = pd.read_html(str(justiceTable),header=1)[0]
df.replace(u'\xc2','',inplace=True)
df.fillna('',inplace=True)
df.columns = ['Justice','State','President','Oath taken','Terminated']

df['Oath taken'] = df['Oath taken'].apply(lambda s: s[3:].strip() if s.find(')')>-1 else s.strip())
df['Terminated'] = df['Terminated'].apply(lambda s: s[3:].strip() if s.find(')')>-1 else s.strip())

df['Oath taken'] = df['Oath taken'].apply(lambda s: s.replace('*',''))
df['Terminated'] = df['Terminated'].apply(lambda s: s.replace('*',''))
```

```python
def read_date(d):
    if len(d)>0:
        d = d.split(',')
        if len(d)==2:
            d = d[0].split(' ')+[d[-1]]
        else:
            d = d[0].split(' ')
        
        try:
            return datetime.strptime(' '.join(d),'%B %d %Y')
        except ValueError:
            print d
    return ''
df['Oath taken'] = df['Oath taken'].apply(lambda d: read_date(d))
df['Terminated'] = df['Terminated'].apply(lambda d: read_date(d))
```

```python
# Split into the Chief and Associate justice tables.
splitix = where(df['Justice']=='Associate Justices')[0][0]
chiefdf = df.iloc[:splitix-2]
assdf = df.iloc[splitix+1:]

assdf.columns = ['Justice','State','President','Oath taken','Terminated']
```
